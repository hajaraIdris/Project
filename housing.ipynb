# **House price prediction using Linear Regression-SingleVariable**

### *Import Libraries*

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

### *Load Dataset from Local Directory*

```python
from google.colab import files
uploaded = files.upload()
```

### *Load Dataset*

```python
dataset = pd.read_csv('dataset.csv')
```

### *Load Summarize*

```python
print(dataset.shape)  # no. of rows and columns
print(dataset.head(5))
```

### *Visualize Dataset*

```python
plt.xlabel('Area')
plt.ylabel('Price')
plt.scatter(dataset['area'], dataset['price'], marker='*')
```

```python
plt.xlabel('Area')
plt.ylabel('Price')
plt.scatter(dataset.area,dataset.price,color='red',marker='+')
```

### *Segregate Dataset into Input X & Output Y*

```python
X = dataset.drop('price',axis='columns')
X
```

```python
Y = dataset.price
Y
```

### *Training Dataset using Linear Regression*

```python
model = LinearRegression()
model.fit(X,Y)
```

### *Predicted Price for Land sq.Feet of custom values*

```python
x=2000
Area=[[x]]
Predicted_y= model.predict(Area)
print(Predicted_y)
```

#**Check if the model is right by verification**
# Theory Calculation
# Y = m * X + b (m is coefficient and b is intercept)

*Coefficient - m*

```python
m=model.coef_
print(m)
```

*Intercept - b*

```python
b=model.intercept_
print(b)
```

### Y=mx+b
*x is Independant variable - Input - area*

```python
y = m*x + b
print("The Price of {0} Square feet Land is: {1}".format(x,y[0]))
```

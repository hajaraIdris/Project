Building visualization

```python
pyplot.bar(df_house['bedrooms'], df_house['price'])
pyplot.xlabel('bedrooms')
pyplot.ylabel('price')
pyplot.title('Bedrooms vs Price in millions')
pyplot.show()
```

```python
pyplot.barh(df_house['bedrooms'], df_house['price'])
pyplot.xlabel('Bedrooms')
pyplot.ylabel('Price in Millions')
pyplot.title('Bedrooms vs Price in Millions')
pyplot.show()
```

```python
# prompt: pie chart showing price on df_house dataframe

prices = df_house['price'].value_counts()
labels = prices.index.to_list()

pyplot.pie(prices, labels=labels, autopct="%1.1f%%")
pyplot.title("Price Distribution in df_house")
pyplot.show()
```

```python
# prompt: pie chart showing total_rooms in  df_house dataframe

rooms = df_house['total_rooms'].value_counts()
labels = rooms.index.to_list()

pyplot.pie(rooms, labels=labels, autopct="%1.1f%%")
pyplot.title("Total Rooms Distribution in df_house")
pyplot.show()
```
